```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\HARSHVARDHAN\Desktop\IPL 2025 DATA_ANALYSIS\IPL2025Bowlers.csv")
```

```python
df.isnull().sum()
```

```python
df["ECO"] = pd.to_numeric(df["ECO"],errors="coerce")
```

```python
df.sort_values("ECO",ascending=False)
```

```python
df.groupby("Team")["WKT"].max()
```

```python
df.sort_values("WKT",ascending=False)
```

```python
total_team_wickets= df.groupby("Team")["WKT"].sum()
total_matches_played = df.groupby("Team")["MAT"].max()

avg_wickets_per_team= total_team_wickets / total_matches_played

for i,v in enumerate(avg_wickets_per_team):
    
    plt.text(i ,v+0.1 , str(round(v)) , ha="center")
    

ax= avg_wickets_per_team.plot(kind = "bar" ,x=total_team_wickets, y=total_matches_played , color="#8680D1")
plt.xlabel("Teams", fontsize=14)
plt.ylabel("Number of wickets ", fontsize= 14)
plt.title("Average wickets per team", fontsize=16)
plt.xticks(rotation=45)
```

```python
total_team_economy =df.groupby("Team")["ECO"].sum()
total_team_matches= df.groupby("Team")["MAT"].max()



avg_economy_per_team= total_team_economy / total_team_matches

for i,v in enumerate(avg_economy_per_team):
    plt.text(i, v +0.1, str(round(v)) , ha= "center")

ax = avg_economy_per_team.plot(kind= "bar",x=total_team_matches, y=total_team_economy , color="#F078AC")
plt.title("Average Bowling Economy Of Team", fontsize=14)
plt.xlabel("Teams" , fontsize=12)
plt.ylabel("Bowling Economy" ,fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
df["ECO"] = pd.to_numeric(df["ECO"], errors="coerce")


top_10_highest_eco= df.nlargest(10 , "ECO")

ax = top_10_highest_eco.plot(kind="bar",x="Player Name" , y="ECO", color="#E1E687")
plt.title("Top 10 Highest Economy of bowlers" , fontsize=16)
plt.xlabel("Players Name" , fontsize=14)
plt.ylabel("Economy of bowlers", fontsize=12)
plt.xticks(rotation=45)
                             
for i,v in enumerate(top_10_highest_eco["ECO"]):
    ax.text(i, v+0.1 , str(v) , ha= "center")
                             
                             
```

```python
top_10_best_eco = df.nsmallest(10, "ECO")

ax = top_10_best_eco.plot(kind="bar", x="Player Name", y="ECO", color="#6CC4A1")
plt.title("Top 10 Most Economical Bowlers", fontsize=16)
plt.xlabel("Player Name", fontsize=12)
plt.ylabel("Economy Rate", fontsize=12)
plt.xticks(rotation=45)
for i, v in enumerate(top_10_best_eco["ECO"]):
    ax.text(i, v + 0.1, str(round(v, 2)), ha="center")
    

plt.tight_layout()

```

```python
top_10_wickets = df.nlargest(10, "WKT")

ax = top_10_wickets.plot(kind="bar", x="Player Name", y="WKT", color="#F7A76C")
plt.title("Top 10 Highest Wicket Takers", fontsize=16)
plt.xlabel("Player Name")
plt.ylabel("Wickets")
plt.xticks(rotation=45)
for i, v in enumerate(top_10_wickets["WKT"]):
    ax.text(i, v + 0.1, str(v), ha="center")

plt.tight_layout()

```
